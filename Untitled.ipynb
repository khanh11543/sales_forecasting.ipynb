```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Dữ liệu sản xuất mẫu
data = {
    'Month': ['2024-01', '2024-02', '2024-03', '2024-04', '2024-05'],
    'Product A': [2000, 2400, 2300, 2500, 2700],
    'Product B': [1800, 2000, 2100, 1900, 2200],
    'Defects': [5, 3, 4, 2, 1],
    'Maintenance_Cost': [300, 280, 250, 260, 240],
    'Customer_Satisfaction': [80, 82, 85, 88, 90]
}
df = pd.DataFrame(data)
df['Total Sales'] = df['Product A'] + df['Product B']
df['Month_Num'] = [1, 2, 3, 4, 5]
```

```python
plt.figure(figsize=(8,4))
sns.lineplot(x=df['Month'], y=df['Total Sales'], marker='o')
plt.title("Biểu đồ 1: Tổng doanh số theo tháng")
plt.xlabel("Tháng")
plt.ylabel("Tổng doanh số")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df, x='Product', y='Sales', estimator=sum)
plt.title("Total Sales by Product")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.lineplot(x=df['Month'], y=df['Product A'], label='Product A', marker='o')
sns.lineplot(x=df['Month'], y=df['Product B'], label='Product B', marker='o')
plt.title("Biểu đồ 2: So sánh doanh số Product A và Product B")
plt.xlabel("Tháng")
plt.ylabel("Doanh số")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.barplot(x=df['Month'], y=df['Defects'], color='salmon')
plt.title("Biểu đồ 3: Số lỗi sản phẩm theo tháng")
plt.xlabel("Tháng")
plt.ylabel("Số lỗi")
plt.tight_layout()
plt.show()

```

```python
plt.figure(figsize=(8,4))
sns.lineplot(x=df['Month'], y=df['Maintenance_Cost'], color='green', marker='o')
plt.title("Biểu đồ 4: Chi phí bảo trì theo tháng")
plt.xlabel("Tháng")
plt.ylabel("Chi phí (USD)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.lineplot(x=df['Month'], y=df['Customer_Satisfaction'], color='purple', marker='o')
plt.title("Biểu đồ 5: Mức độ hài lòng của khách hàng theo tháng")
plt.xlabel("Tháng")
plt.ylabel("Điểm đánh giá (%)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Chuẩn bị dữ liệu đầu vào
X = df[['Month_Num']]  # Đầu vào (biến độc lập)
y = df['Total Sales']  # Đầu ra (biến phụ thuộc)

# Khởi tạo và huấn luyện mô hình
model = LinearRegression()
model.fit(X, y)

# Dự đoán trên tập dữ liệu gốc
y_pred = model.predict(X)

# Dự đoán giá trị cho tháng 6 (Month_Num = 6)
predicted_sales = model.predict(np.array([[6]]))[0]
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='Month_Num', y='Total Sales', data=df, label='Actual Sales', s=100)
plt.plot(df['Month_Num'], y_pred, color='blue', label='Regression Line')
plt.scatter(6, predicted_sales, color='red', label='Forecast June', s=150, marker='X')

plt.title("Biểu đồ 6: Dự báo doanh số tháng 6 bằng Linear Regression")
plt.xlabel("Month (as number)")
plt.ylabel("Total Sales")
plt.legend()
plt.grid(True)
plt.show()
```
